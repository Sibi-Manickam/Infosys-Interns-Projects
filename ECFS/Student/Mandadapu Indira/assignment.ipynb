```python
df = df.rename(columns={'Unnamed: 0': 'DateTime'})
df['DateTime'] = pd.to_datetime(df['DateTime'],format="%d/%m/%Y %H:%M:%S")
df['Date'] = pd.to_datetime(df['DateTime']).dt.date
df.tail()
```

```python
subsetdf = df.copy()
subsetdf = subsetdf.drop('DateTime', axis=1)

subsetdf.set_index('Date', inplace=True)

cumsum_df = subsetdf.cumsum(axis=0)
print(cumsum_df)
```

```python
states = ['Punjab', 'Haryana', 'Rajasthan', 'Delhi', 'UP',
       'Uttarakhand', 'HP', 'J&K', 'Chandigarh', 'Chhattisgarh', 'Gujarat',
       'MP', 'Maharashtra', 'Goa', 'DNH', 'Andhra Pradesh', 'Telangana',
       'Karnataka', 'Kerala', 'Tamil Nadu', 'Pondy', 'Bihar', 'Jharkhand',
       'Odisha', 'West Bengal', 'Sikkim', 'Arunachal Pradesh', 'Assam',
       'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Tripura']
```

```python
count_date = cumsum_df.groupby(cumsum_df.index)[states].sum()

pw_clean = pd.DataFrame(count_date)
pw_clean['Date'] = pd.to_datetime(pw_clean.index)
pw_clean = pw_clean.set_index('Date')

pw_clean.head(20)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
state = 'Delhi'
state_data = pw_clean[[state]]
```

```python
# Check for stationarity
def check_stationarity(timeseries):
    result = adfuller(timeseries)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    for key, value in result[4].items():
        print('Critial Values:')
        print(f'   {key}, {value}')
```

```python
# Plot ACF and PACF
def plot_acf_pacf(timeseries, lags=50):
    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plot_acf(timeseries, lags=lags, ax=plt.gca())
    plt.subplot(122)
    plot_pacf(timeseries, lags=lags, ax=plt.gca())
    plt.show()
```

```python
# Check stationarity and plot ACF/PACF
check_stationarity(state_data[state])
plot_acf_pacf(state_data[state])
```

```python
# Differencing the data to make it stationary if needed
state_data_diff = state_data.diff().dropna()
check_stationarity(state_data_diff[state])
plot_acf_pacf(state_data_diff[state])
```

```python
# Fit SARIMA model
# Adjust the (p,d,q)(P,D,Q,s) parameters based on the ACF/PACF plots and domain knowledge
sarima_model = SARIMAX(state_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_fit = sarima_model.fit(disp=False)
```

```python
# Print model summary
print(sarima_fit.summary())
```

```python
# Forecasting
forecast_steps = 30
forecast = sarima_fit.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=state_data.index[-1], periods=forecast_steps)
forecast_series = pd.Series(forecast.predicted_mean.values, index=forecast_index)
```

```python
# Plotting the forecast
plt.figure(figsize=(15, 8))
plt.plot(state_data, label='Observed')
plt.plot(forecast_series, label='Forecast', linestyle='--')
plt.title(f'{state} Electricity Consumption Forecast')
plt.xlabel('Date')
plt.ylabel('Electricity Consumption')
plt.legend(loc='upper left')
plt.show()
```

```python
# Performance metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Calculate metrics
mae = mean_absolute_error(state_data[state][-forecast_steps:], forecast_series)
mse = mean_squared_error(state_data[state][-forecast_steps:], forecast_series)
rmse = np.sqrt(mse)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
```
